```
import numpy as np
from PIL import Image
```

```
path = os.listdir('.')
path.sort(key=str)
x = np.zeros((2997,224,224,3))
y= np.zeros((2997,12,1))
j = 0
k = 0
for fold in path[:-5]:
  #print(fold)
  path1 = os.listdir(fold)
  path1.sort(key=str)
  for i in path1[:-1]:
    img = Image.open(fold+'/'+i)
    ran = np.random.randint(0,50)
    img = img.crop((ran,ran,224+ran,224+ran))
    img = np.asarray(img)
    if(len(img.shape) != 3):
      continue
    x[j] = img
    label = np.zeros((12,1))
    label[k] = 1
    y[j] = label
    j = j+1
  k = k+1
  print(j)
  print(k)  
#x = np.asarray(x)
#y = np.asarray(y)
```
