```python
import csv
import os
import sys
import os.path
import path
import shutil
from collections import defaultdict

CSVFile = 'F:\\CS\\3-1\\NNFL\\Painters csv\\Van_Gogh.csv'
src = 'F:\\CS\\3-1\\NNFL\\train_1'
dst = 'F:\\CS\\3-1\\NNFL\\Van_Gogh'

with open(CSVFile, 'r') as f:
    reader = csv.reader(f)
    for rowDict in reader:
        #print(rowDict)
        for file in os.listdir(src):
            src_file = os.path.join(src, file)
            dst_file = os.path.join(dst, file)
            if file in rowDict:
                shutil.copyfile(src_file,dst_file)

```

```python
import csv
import os
import sys
import os.path
import path
import shutil
from PIL import Image
import numpy as np
from collections import defaultdict

CSVFile = 'F:\\CS\\3-1\\NNFL\\Van_Gogh\\Van_Gogh.csv'
src = 'F:\\CS\\3-1\\NNFL\\train_1'
dst = 'F:\\CS\\3-1\\NNFL\\Van_Gogh'
dst1 = 'F:\\CS\\3-1\\NNFL\\Van_Gogh_crop'
X = []
with open(CSVFile, 'r') as f:
    reader = csv.reader(f)
    for rowDict in reader:
        img = Image.open(dst+"\\"+rowDict[0])
        img2 = img.crop((0, 0, 224, 224))
        img2.save(dst1+"\\"+rowDict[0], 'JPEG')
              
```
